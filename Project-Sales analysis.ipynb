```python
from collections import Counter
```

```python
sales["COUNTRY"]
```

We have 2823 observations in this series

```python
c=Counter()
for row in sales["COUNTRY"]:
    c.update([row])
c    
```

```python
len(c.keys())
```

So we have 19 Different countries

```python
c.most_common(19)
```

```python
countries=[]
no_of_orders=[]
for i in c.most_common(19):
    countries.append(i[0])
    no_of_orders.append(i[1])    
countries        
```

```python
no_of_orders
```

```python
plt.pie(no_of_orders,labels=countries,wedgeprops={"edgecolor":"black"},autopct="%1.1f%%")
plt.show()
```

# So most orders were from USA with 35.6%

# What are the products ordered and how much they are ordered?

```python
product_counter=Counter()
for row in sales["PRODUCTLINE"]:
    product_counter.update([row])
product_counter    
```

```python
product_type=[]
no_of_products_ordered=[]
for i in product_counter.most_common(7):
    product_type.append(i[0])
    no_of_products_ordered.append(i[1])
product_type
```

```python
no_of_products_ordered
```

```python
plt.pie(no_of_products_ordered,labels=product_type,wedgeprops={"edgecolor":"black"},autopct="%1.1f%%")
plt.show()
```

# Here we have different types of products ordered and how much they are ordered
